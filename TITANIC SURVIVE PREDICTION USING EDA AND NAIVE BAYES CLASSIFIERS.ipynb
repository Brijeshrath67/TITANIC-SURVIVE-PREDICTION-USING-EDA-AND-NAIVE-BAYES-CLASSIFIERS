```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv("/Users/brijesh/Downloads/Code/titanic.csv")    #to read the file
print(df)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
df['Sex'] = le.fit_transform(df['Sex'])
df['Embarked'] = le.fit_transform(df['Embarked'])
```

```python
##Replacing Null
df['Age'] = df['Age'].replace(np.nan, 0)
df['Embarked'] = df['Embarked'].replace(np.nan, 0)
```

```python
df
```

```python
x=df.drop(columns=['Survived','Name','PassengerId','Ticket','Cabin'])
y=df['Survived']      #to create the variable

print("mongo",x)
print("db",y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=12)   #split the value
```

```python
##X_train  - 80% input data
##Y_train  - 80% output data
##X_test   - 20% input data
##Y_test   - 20% output data
```

```python
print("DF",df.shape)
print("x_train",x_train.shape)
print("x_test",x_test.shape)
print("y_train",y_train.shape)
print("y_test",y_test.shape)
```

```python
from sklearn.naive_bayes import GaussianNB  
NB = GaussianNB()
```

```python
NB.fit(x_train, y_train)
```

```python
###train the data
y_pred=NB.predict(x_test)
print("y_pred",y_pred)
print("y_test",y_test)
```

```python
from sklearn.metrics import accuracy_score
print('ACCURACY is', accuracy_score(y_test,y_pred))
```
